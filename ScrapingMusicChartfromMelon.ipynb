# 멜론 인기차트 순위 크롤링

```python
from selenium import webdriver
driver = webdriver.Chrome('C:/Users/CPBGame001/Downloads/chromedriver_win32/chromedriver.exe')
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
songs = soup.select('#lst50')
```

```python
song = songs[8]
```

```python
# 태그명이 div이면서 class 명이 ellipsis이며 rank01인 태그 바로 아래에 있는 span 태그, 그 바로 아래에 존재하는 a태그찾기
title = song.select('div.ellipsis.rank01 > span > a') 
len(title) # 결과 1개
print(title)
```

```python
# 곡 제목 가져오기
title = song.select('div.ellipsis.rank01 > span > a')[0].text
print(title)
```

```python
# 가수 이름 가져오기
artist = song.select('div.ellipsis.rank02 > a')[0].text
print(artist)
```

```python
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    artist = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, '-',artist)
```

```python
# 이상 정리: BS를 활용하여 크롤링하는 버전
# 웹 페이지의 HTMNL정보를 다운로드한 뒤에는 브라우저에서 정보를 가져오지 않음(HTML 다운로드후 브라우저 영향 없음)
# 모든 정보를 다운로드한 뒤 그 안에서 필요한 내용을 찾기에 속도가 빠름
driver = webdriver.Chrome('C:/Users/CPBGame001/Downloads/chromedriver_win32/chromedriver.exe')
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)

from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
songs = soup.select('#lst50')

for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    artist = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, '-',artist)
```

```python
# 셀레늄만 활용하여 크롤링하는 버전
# 웹 페이지에 계속 접속한 상태로 정보를 가져와서 속도가 느림 (웹페이지 연결 유지 필요)
# 클릭,입력 등 조작 가능
driver = webdriver.Chrome('C:/Users/CPBGame001/Downloads/chromedriver_win32/chromedriver.exe')
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)

songs = driver.find_elements_by_css_selector('#lst50')
for song in songs:
    title = song.find_elements_by_css_selector('div.ellipsis.rank01 > span > a')[0].text
    artist = song.find_elements_by_css_selector('div.ellipsis.rank02 > a')[0].text
    print(title,artist, sep='|')
```

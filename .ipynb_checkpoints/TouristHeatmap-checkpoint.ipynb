# 데이터 전처리 과정을 함수로 만들기

```python
def create_kto_data(yy, mm):
    file_path = '../Downloads/datasalon-master/datasalon-master/3_Tourists_Event/files/kto_{}{}.xlsx'.format(yy,mm)
    df = pd.read_excel(file_path, header = 1, skipfooter = 4, usecols = 'A:G')
    df['기준년월'] = '{}-{}'.format(yy,mm)
    ignore_list = ['아시아주', '미주', '구주','대양주','아프리카주','기타대륙','교포소계']
    condition = (df.국적.isin(ignore_list) == False)
    df_country = df[condition].reset_index(drop=True)
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']
    df_country['대륙'] = continents
    df_country['관광객비율(%)'] = round(df_country.관광 / df_country.계*100,1)
    tourist_sum = sum(df_country['관광'])
    df_country['전체비율(%)'] = round(df_country['관광'] / tourist_sum * 100,1)
    return(df_country)
```

```python
kto_test = create_kto_data(2018, 12)
kto_test
```

```python
df = pd.DataFrame()
```

```python
# 새로 만든 빈 데이터프레임에 2010년부터 2020년까지 반복문으로 데이터 추가하기. 
# 오류가 나는데 2019년 09월부터 자료가 없기 때문. 있는 데이터까지는 데이터 추가 잘 됨

for yy in range(2010, 2020):
    for mm in range(1, 13):
        temp = create_kto_data(str(yy), str(mm).zfill(2))
        df = df.append(temp, ignore_index = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 태국 국적 데이터 필터링
condition = (df.국적 == '태국')
df_thai = df[condition]
df_thai
```

```python
# 월별 그래프
plt.figure(figsize=(20,8))
plt.plot(df_thai['기준년월'], df_thai['관광'])
plt.title('타이 국적의 관광객 월별 추이')
plt.xlabel('기준년월'); plt.ylabel('관광객수')
plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01'])
plt.show()
```

```python
df_thai['년도'] = df_thai['기준년월'].str.slice(0,4)
df_thai['월'] = df_thai['기준년월'].str.slice(5,7)
```

```python
df_pivot = df_thai.pivot_table(values='관광', index = '년도', columns = '월')
df_pivot
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r') # annot: 각 칸에 수치 표시, fmt='.0f': 소수점 없이 실수 표시
plt.title('타이 국적의 관광객 월별 히트맵')
plt.show()
```
